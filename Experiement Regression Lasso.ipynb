# Analysis by Systems 

### Build Borda Count

```python
considered_df_system = pd.read_csv(os.path.join(data_path,ds)).set_index(['System','Utterance'])
assert considered_df_system.shape[0] % considered_df_system.loc[0].shape[0] == 0
nbs_system = considered_df_system.shape[0] // considered_df_system.loc[0].shape[0] 
nbs_utt = considered_df_system.loc[0].shape[0] 
```

```python
considered_df_system
```

```python
dic = {}

for column in list(considered_df_system.columns):
    print(column)
    dic['borda_{}'.format(column)] = []
    for i in range(nbs_system): 
        current_borda_count = np.argsort(np.argsort(considered_df_system.loc[i][column].values)).tolist()
        dic['borda_{}'.format(column)] += current_borda_count
dic['System'] = sum([['M{}'.format(i)]*nbs_utt for i in range(nbs_system)],[])
dic['Utt'] = sum([['U{}'.format(i) for i in range(nbs_utt)]* nbs_system],[])
```

```python
dic['System'] 
```

```python
df_borda_system = pd.DataFrame(dic).set_index('System')
df_borda_system.head()
human_metric = [ i for i in df_borda_system.columns if ('H:' in i and 'Utt' not in i)]
other_metric = [ i for i in df_borda_system.columns if ('H:' not in i and 'Utt' not in i)]
print("Human", human_metric)
print("Other", other_metric)
```

```python
df_borda_system
```

### Use the automatic metrics to predict one judgement

```python
X_train = df_borda_system[other_metric].values
y_train = df_borda_system[human_metric].values
X_train.shape,y_train.shape,y_train[:,0].shape
```

```python
final_resulting_dic_residual = {
    'metric':[],
    'alpha':[],
    'mse':[]
}
```

```python

index_metric = 0
for index_metric in range(0,len(human_metric)):
    res = {'coeff': [],"intercept": [],"alpha": [],"mse": [],"r2": []}
    for alpha in [0] +list(np.logspace(-2, 2.3, num=20)) : #range(0,100,10):
        if alpha == 0:
            clf = linear_model.LinearRegression(positive = False)
        else: 
            clf = linear_model.Lasso(alpha=alpha,positive = False)
        clf.fit(X_train,y_train[:,index_metric])
        res['coeff'].append(clf.coef_.tolist())
        res['alpha'].append(alpha)
        res['intercept'].append(clf.intercept_)
        
        y_pred = clf.predict(X_train)
        mse = mean_squared_error(y_train[:,index_metric], y_pred) / max(y_train[:,index_metric])
        r2 = r2_score(y_train[:,index_metric], y_pred)
        res['mse'].append(mse)
        res['r2'].append(r2)
        print("Mean squared error: %.2f" % mse)
        print("Coefficient of determination: %.2f" % r2 )
    target_metric = human_metric[index_metric].replace(':','')
    
    for index,metric in enumerate(other_metric):
        for index_alpha,l in enumerate(res['mse']):
            
            final_resulting_dic_residual['mse'].append(l)
            final_resulting_dic_residual['alpha'].append(res['alpha'][index_alpha])
            final_resulting_dic_residual['metric'].append(f'w_{target_metric}')
    resulting_dic = {
        'Metric':[],
        'alpha':[],
        'coeff':[]
    }
    for index,metric in enumerate(other_metric):
        for index_alpha,l in enumerate(res['coeff']):
            resulting_dic['coeff'].append(l[index])
            resulting_dic['Metric'].append(metric)
            resulting_dic['alpha'].append(res['alpha'][index_alpha])
    fig,ax = plt.subplots(figsize=(11.7,8.27))
    line_df = pd.DataFrame(resulting_dic)
    line_df.Metric = line_df.Metric.str.replace("borda_", "")
    ax = sns.lineplot(data=line_df, x="alpha", y="coeff", hue="Metric", style="Metric",markers=True, markersize=15)
    ax.set_xscale('log')

    plt.setp(ax.get_xticklabels(), fontsize=25,ha='center')
    plt.setp(ax.get_yticklabels(), fontsize=25)

    lgd = plt.legend(prop={'size': 15})
    for i in range(len(lgd.legendHandles)):
        lgd.legendHandles[i]._legmarker.set_markersize(15)
    plt.xlabel('$\\alpha$', fontsize=40)
    plt.ylabel('Coeff', fontsize=40)
    plt.tight_layout()
    metric_name = human_metric[index_metric].replace(':','')
    plt.savefig(f'coeff/{ds}_{metric_name}_system.pdf', format = 'pdf')
    plt.close()
```
